```python
import pandas as pd
import folium as fl

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
country_geo_f = 'countries.geo.json'
country_pop_f = 'population-percentages.csv'

test_json = 'test.json'
```

```python
country_pop = pd.read_csv(country_pop_f)
country_pop.head(2)
```

```python
test_map = fl.Map(location=[0, 0], zoom_start=2)
test_map.choropleth(geo_path=country_geo_f, # the map file
                   data=country_pop, # the attributes DF
                   columns=['Country_Code', 'Total population perc 35-39'], # attributes
                   key_on='feature.id',
                   fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2)
```
